```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def manifold_plot_3d(df, x_col_name, y_col_name, z_col_name, bins=20, x_min=None, x_max=None, y_min=None, y_max=None):
    X_COL_NAME= x_col_name
    X_COL_BIN = f'{X_COL_NAME}_binned'
    Y_COL_NAME = y_col_name
    Y_COL_BIN = f'{Y_COL_NAME}_binned'
    Z_COL_NAME = z_col_name 
    # Assuming df is your DataFrame
    # Group by 'a_n' and 'b_n' and calculate the mean of 'fraction_difference_signal_from_distribution_in_ci'
    df[X_COL_BIN] = pd.cut(df[X_COL_NAME], bins=bins)
    df[Y_COL_BIN] = pd.cut(df[Y_COL_NAME], bins=bins)
    grouped_df = df.groupby([X_COL_BIN, Y_COL_BIN], observed=False)[Z_COL_NAME].mean().reset_index()

    # Get bin centers
    x_center_name = f'{X_COL_NAME}_center'
    y_center_name = f'{Y_COL_NAME}_center'
    grouped_df[x_center_name] = grouped_df[X_COL_BIN].apply(lambda x: x.mid)
    grouped_df[y_center_name] = grouped_df[Y_COL_BIN].apply(lambda x: x.mid)
    grouped_df[x_center_name] = pd.to_numeric(grouped_df[x_center_name])
    grouped_df[y_center_name] = pd.to_numeric(grouped_df[y_center_name])


    # Filter based on x_min, x_max, y_min, y_max
    if x_min is not None:
        grouped_df = grouped_df[grouped_df[x_center_name] >= x_min]
    if x_max is not None:
        grouped_df = grouped_df[grouped_df[x_center_name] <= x_max]
    if y_min is not None:
        grouped_df = grouped_df[grouped_df[y_center_name] >= y_min]
    if y_max is not None:
        grouped_df = grouped_df[grouped_df[y_center_name] <= y_max]
    

    # Pivot the DataFrame to create a grid
    pivot_df = grouped_df.pivot(index=x_center_name, columns=y_center_name, values=Z_COL_NAME)

    # Create meshgrid for a_n and b_n
    a_n = pivot_df.index.values
    b_n = pivot_df.columns.values
    A_n, B_n = np.meshgrid(a_n, b_n)

    # Get the Z values from the pivot table
    Z = pivot_df.values.T  # Transpose to match the meshgrid shape

    # Create a 3D plot
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Surface plot
    surf = ax.plot_surface(A_n, B_n, Z, cmap='viridis')

    # Set axis labels
    ax.set_xlabel(X_COL_NAME)
    ax.set_ylabel(Y_COL_NAME)
    ax.set_zlabel(Z_COL_NAME)

    # Add colorbar
    fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

    # Add text labels (optional, might be crowded with 25 bins)
    lenbn = len(b_n)
    lenan = len(a_n)
    for i in range(lenbn):
        for j in range(lenan):
            if i % int(lenbn/6 + 1) == 0 and j % int(lenan/6 + 1) == 0:  # Reduced frequency of labels
                ax.text(A_n[i, j], B_n[i, j], Z[i, j], f'({A_n[i,j]:.0f}, {B_n[i,j]:.0f}, {Z[i, j]:.2f})', color='black', fontsize=8)
            #ax.text(A_n[i, j], B_n[i, j], Z[i, j], f'{Z[i, j]:.2f}', color='black', fontsize=8)

    ax.view_init(30, 220)
    # move the z label in
    ax.zaxis.labelpad = -5

    plt.tight_layout()
    plt.title(f'{Z_COL_NAME} as a function of {X_COL_NAME} and {Y_COL_NAME}')
    plt.show()

#manifold_plot_3d(df, 'au_n', 'bu_n', 'fraction_difference_signal_from_distribution_in_ci', bins=25)
#manifold_plot_3d(df, 'au_n', 'bu_n', 'day_diff_sensitivity', bins=18)
#manifold_plot_3d(df, 'au_n', 'bu_n', 'day_diff_specific', bins=30)

df['scale_diff'] = df['distribution_a_scale'] - df['distribution_b_scale']
df['loc_diff'] = df['distribution_a_loc'] - df['distribution_b_loc']
#manifold_plot_3d(df, 'scale_diff', 'loc_diff', 'fraction_difference_signal_from_distribution_in_ci', bins=25, x_min=-50, x_max=50, y_min=-25, y_max=25)
manifold_plot_3d(df, 'distribution_a_loc', 'distribution_b_loc', 'fraction_difference_signal_from_distribution_in_ci')
```
