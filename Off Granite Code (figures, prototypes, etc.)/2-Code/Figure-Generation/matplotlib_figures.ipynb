```python
def generate_combined_hospital_results_plot(data1, data2):
    """
    Generate a figure with two hospital results side by side, each with diff signal plot
    above diff signal stats plot. Y-axes are aligned between subplots.
    
    Parameters:
    -----------
    data1, data2 : dict
        Dictionaries containing the required plotting data for each hospital
    """
    import matplotlib.pyplot as plt
    import numpy as np
    
    figsize = (4.9, (3/4)*4.9)  # doubled width from original
    plt.rcParams.update({'font.size': 8})
    
    def plot_threshold_regions(ax, x, mask, color, alpha):
        start_idx = None
        for i in range(len(mask)):
            if mask[i] and start_idx is None:
                start_idx = i
            elif not mask[i] and start_idx is not None:
                ax.fill_between(x[start_idx:i], 
                              -1 * np.ones(i-start_idx),
                              1 * np.ones(i-start_idx),
                              color=color, alpha=alpha)
                start_idx = None
    
        if start_idx is not None:
            ax.fill_between(x[start_idx:], 
                          -1 * np.ones(len(x)-start_idx),
                          1 * np.ones(len(x)-start_idx),
                          color=color, alpha=alpha)
    
    # Create figure with four subplots arranged in a 2x2 grid
    fig = plt.figure(figsize=figsize, constrained_layout=True)
    gs = fig.add_gridspec(2, 2, height_ratios=[1, 1], hspace=0.00, wspace=0.0)
    
    # Create subplots
    ax1_left = fig.add_subplot(gs[0, 0])
    ax1_right = fig.add_subplot(gs[0, 1])
    ax2_left = fig.add_subplot(gs[1, 0])
    ax2_right = fig.add_subplot(gs[1, 1])
    
    # Process both datasets
    for data, ax1, ax2, title in [(data1, ax1_left, ax2_left, 'Hospital 1'),
                                 (data2, ax1_right, ax2_right, 'Hospital 2')]:
        
        x = data['day_ind_on_0_199']
        x = 199 - np.array(x)
        
        colors = get_longitudinal_care_delivery_paper_colormap()
        
        # Top plot (diff signal)
        ax1.plot(x, data['diffSignal'], 
                color=colors['diffSignal'], 
                linewidth=2, 
                label='difference signal')
        
        ax1.fill_between(x, 
                        data['cilow'], 
                        data['cihigh'], 
                        color=colors['ci'], 
                        alpha=0.2, 
                        label='95% credible interval')
        
        # Bottom plot (diff signal stats)
        ax2.plot(x, data['dbdfe'], 
                color=colors['dbdfe'], 
                linewidth=2, 
                label='dbdfe')
        
        ax2.plot(x, data['wmpcd'], 
                color=colors['wmpcd'], 
                linewidth=2, 
                label='wmpcd')
        
        ax2.plot(x, data['pmpcd'], 
                color=colors['pmpcd'], 
                linewidth=2, 
                label='pmpcd')
        
        # Create masks for threshold regions
        wmpcd_threshold_mask = [w > 0.95 for w in data['wmpcd']]
        pmpcd_threshold_mask = [p > 0.95 for p in data['pmpcd']]
        dbdfe_threshold_mask = [d < 0.05 for d in data['dbdfe']]
        
        # Plot threshold regions
        for ax in [ax1, ax2]:
            plot_threshold_regions(ax, x, wmpcd_threshold_mask, color=colors['wmpcd'], alpha=0.3)
            plot_threshold_regions(ax, x, pmpcd_threshold_mask, color=colors['pmpcd'], alpha=0.3)
            plot_threshold_regions(ax, x, dbdfe_threshold_mask, color=colors['dbdfe'], alpha=0.15)
        
        # Customize plots
        # ax1.set_ylim(min(data['cilow']) - 0.02, max(data['cihigh']) + 0.02)
        ax1.set_ylim(-0.4, 0.4)
        ax2.set_ylim(0, 1)
        
        ax1.axhline(0, color='black', linewidth=0.5)
        
        print(x.max(), x.min())
        ax1.set_xlim(27, -1)
        ax2.set_xlim(27, -1)
        
        for ax in [ax1, ax2]:
            ax.grid(True, alpha=0.5, linewidth=0.2, color='black')
            ax.spines['top'].set_visible(False)
            ax.spines['right'].set_visible(False)
            
        # Add title
        ax1.set_title(title)
    
    # Create patches for legend
    from matplotlib.patches import Patch
    wm_patch = Patch(color=colors['wmpcd'], alpha=0.3, label='wmpcd > 0.95')
    pm_patch = Patch(color=colors['pmpcd'], alpha=0.3, label='pmpcd > 0.95')
    # dbdfe_patch = Patch(color=colors['dbdfe'], alpha=0.15, label='dbdfe < 0.05')
    
    # Add labels (only for left plots and bottom plots)
    ax2_left.set_xlabel('Days Before Death')
    ax2_right.set_xlabel('Days Before Death')
    ax1_left.set_ylabel('$\mathbf{diffSignal}$')
    ax2_left.set_ylabel('Probability')
    
    # Align y-labels
    fig.align_ylabels([ax1_left, ax2_left])
    
    return fig

# Example usage:
with open('../../1-Input/Figure-and-Paper-Data/example_hospital_results_390111.json') as f:
    data1 = json.load(f)
with open('../../1-Input/Figure-and-Paper-Data/example_hospital_results_450044.json') as f:
    data2 = json.load(f)

fig = generate_combined_hospital_results_plot(data1, data2)
fig.savefig('../../3-Output/GeneratedFigures/combined_hospital_results.jpg', dpi=600)
plt.show()
```

```python
import json
import os
import textwrap
import typing
import pyphen
from typing import Any, Dict, List

from matplotlib.figure import Figure
import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

# --- Type Hinting for Clarity ---
HospitalData = Dict[str, Any]

# =============================================================================
# 1. HELPER AND PLOTTING FUNCTIONS
# =============================================================================

def wrap_with_hyphenation(text, width):
    dic = pyphen.Pyphen(lang='en')
    
    lines = []
    words = text.split()
    current_line = ""
    
    for word in words:
        # Try to add the word to current line
        test_line = current_line + (" " if current_line else "") + word
        
        if len(test_line) <= width:
            current_line = test_line
        else:
            # Word doesn't fit - try to hyphenate it
            if current_line:  # If line has content
                space_left = width - len(current_line) - 1
                
                # Get all possible hyphenation points
                hyphenated_word = dic.inserted(word, hyphen='-')
                
                # Find the longest prefix that fits
                best_split = None
                for pos in range(len(hyphenated_word)):
                    if hyphenated_word[pos] == '-' and pos <= space_left:
                        best_split = pos
                
                if best_split:
                    # Split the word at the hyphen
                    prefix = hyphenated_word[:best_split]
                    suffix = hyphenated_word[best_split+1:]  # Skip the hyphen
                    
                    lines.append(current_line + " " + prefix)
                    current_line = suffix
                else:
                    # No good hyphenation point, move whole word to next line
                    lines.append(current_line)
                    current_line = word
            else:
                current_line = word
    
    if current_line:
        lines.append(current_line)
    
    return lines

def plot_data_panel(ax: plt.Axes, data: HospitalData, colors: Dict, plot_type: str) -> None:
    """A single function to draw any of the three data panel types."""
    import numpy as np # Ensure numpy is available
    x = 199 - np.array(data['day_ind_on_0_199'])
    if plot_type == 'diffSignal':
        ax.plot(x, data['diffSignal'], color=colors['diffSignal'], lw=1.5)
        ax.fill_between(x, data['cilow'], data['cihigh'], color=colors['ci'], alpha=0.3)
        ax.axhline(0, color='black', lw=0.5, ls='--')
        ax.set_ylim(-0.4, 0.4)
    elif plot_type == 'probability':
        ax.plot(x, data['pmpcd'], color=colors['pmpcd'], lw=1.5)
        ax.plot(x, data['wmpcd'], color=colors['wmpcd'], lw=1.5)
        ax.fill_between(x, 0, 1, where=np.array(data['pmpcd']) > 0.95, color=colors['pmpcd'], alpha=0.3, step='mid')
        ax.fill_between(x, 0, 1, where=np.array(data['wmpcd']) > 0.95, color=colors['wmpcd'], alpha=0.3, step='mid')
        ax.set_ylim(-0.05, 1.05)
    elif plot_type == 'dbdfe':
        ax.plot(x, data['dbdfe'], color=colors['dbdfe'], lw=1.5)
        ax.axhline(0.05, color=colors['dbdfe'], ls='--', lw=0.8)
        ax.fill_between(x, 0, 1, where=np.array(data['dbdfe']) < 0.05, color=colors['dbdfe'], alpha=0.15, step='mid')
        ax.set_ylim(-0.05, 1.05)
    
    ax.set_xlim(max(x), -1)
    ax.grid(True, axis='y', which='major', linestyle=':', alpha=0.6)

def create_div_like_text_panel(ax, title_text, description_text, result1_text, result2_text):
    """
    Creates a div-like layout in a matplotlib axes with three rows:
    - Row 1: Title/Section label
    - Row 2: Description
    - Row 3: Results (two lines)
    """
    ax.axis('off')  # Hide axes
    
    # Helper function to get text height
    def get_text_height(text_obj):
        ax.figure.canvas.draw()
        bbox = text_obj.get_window_extent(renderer=ax.figure.canvas.get_renderer())
        bbox_axes = bbox.transformed(ax.transAxes.inverted())
        return bbox_axes.height
    

    # Starting position and spacing
    current_y = 1 - 0.05  # Start from top with margin
    vertical_spacing = 0.03  # Space between sections
    
    # Row 1: Title/Section Label (top)
    title_obj = ax.text(0.00, current_y, title_text,
                       ha='left', va='top', fontsize=9, weight='bold',
                       transform=ax.transAxes)
    
    # Calculate next position based on title height
    title_height = get_text_height(title_obj)
    current_y -= (title_height + vertical_spacing)
    
    # Row 2: Description (middle)
    #description_lines = textwrap.wrap(hyphenate_text(description_text), width=40)
    description_lines = textwrap.wrap(description_text, width=40)
    description_text_wrapped = "\n".join(description_lines)
    
    description_obj = ax.text(0.00, current_y, description_text_wrapped,
                             ha='left', va='top', fontsize=9,
                             transform=ax.transAxes)
    
    # Calculate next position based on description height
    description_height = get_text_height(description_obj)
    current_y -= (description_height + vertical_spacing)
    
    # Row 3: Results (bottom) - two separate lines
    result1_obj = ax.text(0.00, current_y, result1_text,
                         ha='left', va='top', fontsize=9,
                         transform=ax.transAxes)
    
    # Position second result line below the first
    result1_height = get_text_height(result1_obj)
    current_y -= (result1_height + 0.01)  # Smaller spacing between result lines
    
    result2_obj = ax.text(0.00, current_y, result2_text,
                         ha='left', va='top', fontsize=9,
                         transform=ax.transAxes)

# def create_div_like_text_panel(ax, title_text, description_text, result1_text, result2_text):
#        """
#        Creates a div-like layout in a matplotlib axes with three rows:
#        - Row 1: Title/Section label
#        - Row 2: Description 
#        - Row 3: Results (two lines)
#        """
#        ax.axis('off')  # Hide axes
   
#        # Row 1: Title/Section Label (top)
#        ax.text(0.00, 1 - 0.05, title_text, 
#                ha='left', va='top', fontsize=9, weight='bold',
#                transform=ax.transAxes)
   
#        # Row 2: Description (middle)
#        description_lines = textwrap.wrap(description_text, width=40)
#        num_lines = len(description_lines)
#        description_text = "\n".join(description_lines)
#        ax.text(0.00, 0.85 - 0.05 - 0.03, description_text,
#                ha='left', va='top', fontsize=9,
#                transform=ax.transAxes, wrap=True)
   
#        # Row 3: Results (bottom) - two separate lines
#        ax.text(0.00, 0.50 - 0.06 * num_lines, result1_text,
#                ha='left', va='top', fontsize=9,
#                transform=ax.transAxes)
   
#        ax.text(0.00, 0.50 - 0.05 - 0.06 * num_lines, result2_text,
#                ha='left', va='top', fontsize=9,
#                transform=ax.transAxes)

# def create_div_like_text_panel(ax: plt.Axes, title_text: str, description_text: str, result1_text: str, result2_text: str) -> None:
#     """
#     Creates a div-like layout in a matplotlib axes with three rows,
#     preserving your specified structure and formatting.
#     """
#     ax.axis('off')  # Hide axes
    
#     # Row 1: Title using mathtext for bolding
#     ax.text(0.00, 1, fr"$\bf{{{title_text}}}$", 
#             ha='left', va='top', fontsize=9,
#             transform=ax.transAxes)

#     # Row 2: Description, wrapped
#     description_lines = textwrap.wrap(description_text, width=45)
#     description_text_wrapped = "\n".join(description_lines)
#     ax.text(0.00, 0.80, description_text_wrapped,
#             ha='left', va='top', fontsize=9,
#             transform=ax.transAxes, wrap=True)

#     # Row 3: Results
#     ax.text(0.00, 0.45, result1_text,
#             ha='left', va='top', fontsize=9,
#             transform=ax.transAxes)
#     ax.text(0.00, 0.30, result2_text,
#             ha='left', va='top', fontsize=9,
#             transform=ax.transAxes)

# =============================================================================
# 3. MAIN FIGURE ASSEMBLY
# =============================================================================

def create_master_figure(data1: HospitalData, data2: HospitalData) -> Figure:
    """Constructs the complete figure using the user-provided layout."""
    # --- Font and Style Setup ---
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'DejaVu Sans'],
        'font.size': 8
    })
    colors = get_longitudinal_care_delivery_paper_colormap()
    
    # --- Figure and Layout Definition (from your code) ---
    fig = plt.figure(figsize=(7, 6.5))

    # Container 1: Top row for text panels
    gs_top = fig.add_gridspec(
        1, 3, 
        top=1, bottom=0.8, 
        left=0.00, right=1,
        wspace=0.00

    )

    # Container 2: Main grid for plots and legends
    gs_main = fig.add_gridspec(
        3, 3, 
        top=0.775, bottom=0.05,
        left=0.06, right=0.92,
        height_ratios=[1, 1, 1], 
        width_ratios=[4, 4, 2.5],
        wspace=0.1, hspace=0.1
    )

    # --- Populate Top Panels ---
    ax_a = fig.add_subplot(gs_top[0, 0])
    create_div_like_text_panel(
        ax_a,
        title_text="a. Fisher Exact ($\\mathbf{FE}$) Test",
        description_text="Tests if overall hospice utilization rates differ between racial groups",
        result1_text=r"H1: $p = 0.077 \geq 0.05\ \therefore\ \mathbf{FE}(-)$",
        result2_text=r"H2: $p = 0.046 < 0.05\ \therefore\ \mathbf{FE}(+)$"
    )

    ax_b = fig.add_subplot(gs_top[0, 1])
    create_div_like_text_panel(
        ax_b,
        title_text="b. First Day of Hospice($\\mathbf{FDH}$) Test",
        description_text="Tests if timing of first hospice day differs between racial groups for pt. who received hospice using a Mann-Whitney U test",
        result1_text=r"H1: $p = 0.010 < 0.05\ \therefore\ \mathbf{MW}(+)$",
        result2_text=r"H2: $p = 0.288 \geq 0.05\ \therefore\ \mathbf{MW}(-)$"
    )
    
    ax_c = fig.add_subplot(gs_top[0, 2])
    create_div_like_text_panel(
        ax_c,
        title_text="c. Number of Hospice Days ($\\mathbf{NHD}$) Test",
        description_text="Tests if number of days in hospice differs between racial groups for all pts. using a Mann-Whitney U test",
        result1_text=r"H1: $p = 0.052 \geq 0.05\ \therefore\ \mathbf{NHD}(-)$",
        result2_text=r"H2: $p = 0.050 \geq 0.05\ \therefore\ \mathbf{NHD}(-)$"
    )

    # --- Populate Plot and Legend Grid ---
    axes_info = {
        'd': {'plot_type': 'diffSignal', 'ylabel': r'$\mathbf{diffSignal}$'},
        'e': {'plot_type': 'probability', 'ylabel': r'Probability'},
        'f': {'plot_type': 'dbdfe', 'ylabel': r'$\mathbf{DBDFE}$ $p$-value'},
    }
    
    legend_handles = {
        'd': [mlines.Line2D([], [], c=colors['diffSignal'], label=r'$\mathbf{diffSignal}$'),
              mpatches.Patch(fc=colors['ci'], alpha=0.3, label='Pointwise 95% \nCredible Interval')],
        'e': [mlines.Line2D([], [], c=colors['pmpcd'], label=r'$\mathbf{PMPCD}$: $p(\mathrm{diffSignal} < -0.05)$'),
              mpatches.Patch(fc=colors['pmpcd'], alpha=0.3, label=r'$\mathbf{PMPCD(+)}: > 0.95$'),
              mlines.Line2D([], [], c=colors['wmpcd'], label=r'$\mathbf{WMPCD}$: $p(\mathrm{diffSignal} > 0.05)$'),
              mpatches.Patch(fc=colors['wmpcd'], alpha=0.3, label=r'$\mathbf{WMPCD(+)}: > 0.95$')],
        'f': [mlines.Line2D([], [], c=colors['dbdfe'], label=r'$\mathbf{DBDFE}$ $p$-value'),
              mpatches.Patch(fc=colors['dbdfe'], alpha=0.2, label=r'$\mathbf{DBDFE(+)}: < 0.05$')]
    }

    left_axes = []
    for row, (panel_letter, info) in enumerate(axes_info.items()):
        # Create plot axes
        ax1 = fig.add_subplot(gs_main[row, 0])
        ax2 = fig.add_subplot(gs_main[row, 1], sharey=ax1)
        if row > 0: ax1.sharex(left_axes[0])
        if row > 0: ax2.sharex(fig.axes[-(3+2)]) # Share with the plot above it in the same column
        
        # Plot data
        plot_data_panel(ax1, data1, colors, plot_type=info['plot_type'])
        plot_data_panel(ax2, data2, colors, plot_type=info['plot_type'])
        
        # Style axes
        ax1.set_ylabel(info['ylabel'])
        ax1.text(-0.2, 1.00, panel_letter+".", transform=ax1.transAxes, ha='left', va='top', fontsize=9, fontweight='bold')

        if row == 0:
            ax1.set_title("Hospital 1", fontweight='bold')
            ax2.set_title("Hospital 2", fontweight='bold')
        if row < 2:
            ax1.tick_params(axis='x', labelbottom=False)
            ax2.tick_params(axis='x', labelbottom=False)
        else:
            ax1.set_xlabel("Days Before Death")
            ax2.set_xlabel("Days Before Death")
            
        ax2.tick_params(axis='y', labelleft=False)
        left_axes.append(ax1)
        
        # Create legend
        ax_leg = fig.add_subplot(gs_main[row, 2]); ax_leg.axis('off')
        ax_leg.legend(handles=legend_handles[panel_letter], loc='center left', frameon=False)
    
    # Add to your existing figure
    #fig.text(0.5, 0.5, 'PLACEHOLDER', 
         #fontsize=50, color='red', alpha=0.7,
         #ha='center', va='center', 
         #transform=fig.transFigure,
         #rotation=45, weight='bold')
        
    fig.align_ylabels(left_axes)
    return fig

# =============================================================================
# 4. SCRIPT EXECUTION
# =============================================================================
if __name__ == '__main__':
    try:
        # NOTE: Ensure these paths are correct relative to where you run the script
        data_path_1 = '../../1-Input/Figure-and-Paper-Data-Resubmit/example_hospital_results_390111_0.05.json'
        data_path_2 = '../../1-Input/Figure-and-Paper-Data-Resubmit/example_hospital_results_450044_0.05.json'
        
        # Lazy import of numpy since it's only needed here
        import numpy as np
        with open(data_path_1) as f: hospital_data_1 = json.load(f)
        with open(data_path_2) as f: hospital_data_2 = json.load(f)

        final_figure = create_master_figure(hospital_data_1, hospital_data_2)
        
        output_dir = '../../3-Output/GeneratedFigures/'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        output_basename = os.path.join(output_dir, 'final_figure_from_user_layout')
        final_figure.savefig(f'{output_basename}.pdf', bbox_inches='tight')
        final_figure.savefig(f'{output_basename}.png', dpi=300, bbox_inches='tight')

        print(f"Successfully generated and saved FINAL figure to:\n{output_basename}.pdf\n{output_basename}.png")
        plt.show()

    except FileNotFoundError as e:
        print(f"--- ERROR: DATA FILE NOT FOUND ---\n{e.filename}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
# -*- coding: utf-8 -*-
"""
Final, Refactored Code for Generating a Multi-Panel Hospital Data Figure.

This script demonstrates a best-practice approach to creating complex,
publication-quality figures with Matplotlib, emphasizing clarity, modularity,
and maintainability.

Design Principles:
- Separation of Concerns: Style, layout, and plotting logic are decoupled.
- Modularity: Each component of the figure is drawn by a dedicated function.
- Readability: Clear naming, type hinting, and comments make the code self-documenting.
"""

import json
import os
import typing
from typing import Any, Dict, List

import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.gridspec import GridSpec

# --- Type Hinting for Clarity ---
HospitalData = Dict[str, Any]
AxesTuple = typing.Tuple[plt.Axes, plt.Axes, plt.Axes]


# =============================================================================
# 1. STYLE AND CONFIGURATION
# =============================================================================

def get_color_palette() -> Dict[str, str]:
    """Returns the color palette for the figure."""
    return {
        'diffSignal': '#0072B2',
        'ci': '#D55E00',
        'pmpcd': '#CC79A7',
        'wmpcd': '#009E73',
        'dbdfe': '#56B4E9',
        'significance': '#D40E15'
    }

def setup_matplotlib_style() -> None:
    """Configures Matplotlib's rcParams for a consistent, professional appearance."""
    plt.rcParams.update({
        "text.usetex": True,
        "font.family": "sans-serif",
        "text.latex.preamble": r"""
            \usepackage{amsmath}
            \usepackage{amssymb}
            \usepackage[T1]{fontenc}
            \usepackage{helvet}
            \renewcommand\familydefault{\sfdefault}
        """,
        "font.sans-serif": ["Helvetica"],
        'font.size': 8
    })

# =============================================================================
# 2. MODULAR PLOTTING FUNCTIONS
# =============================================================================

def draw_text_panel(ax: plt.Axes, content: str) -> None:
    """Draws a formatted text block onto a given axis."""
    ax.axis('off')
    ax.text(0, 0.5, content, ha='left', va='center', fontsize=9, wrap=True)

def plot_diffsignal_panel(ax: plt.Axes, data: HospitalData, colors: Dict[str, str]) -> None:
    """Draws the diffSignal panel (d)."""
    x = 199 - np.array(data['day_ind_on_0_199'])
    ax.plot(x, data['diffSignal'], color=colors['diffSignal'], linewidth=1.5)
    ax.fill_between(x, data['cilow'], data['cihigh'], color=colors['ci'], alpha=0.3)
    ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
    ax.set_ylim(-0.4, 0.4)

def plot_probability_panel(ax: plt.Axes, data: HospitalData, colors: Dict[str, str]) -> None:
    """Draws the PMPCD/WMPCD probability panel (e)."""
    x = 199 - np.array(data['day_ind_on_0_199'])
    ax.plot(x, data['pmpcd'], color=colors['pmpcd'], linewidth=1.5)
    ax.plot(x, data['wmpcd'], color=colors['wmpcd'], linewidth=1.5)
    
    pmpcd_mask = np.array(data['pmpcd']) > 0.95
    wmpcd_mask = np.array(data['wmpcd']) > 0.95
    ax.fill_between(x, 0, 1, where=pmpcd_mask, facecolor=colors['pmpcd'], alpha=0.3, step='mid')
    ax.fill_between(x, 0, 1, where=wmpcd_mask, facecolor='none', edgecolor=colors['wmpcd'], hatch='//', alpha=0.7, step='mid')
    ax.set_ylim(-0.05, 1.05)
    
def plot_dbdfe_panel(ax: plt.Axes, data: HospitalData, colors: Dict[str, str]) -> None:
    """Draws the DBDFE p-value panel (f)."""
    x = 199 - np.array(data['day_ind_on_0_199'])
    ax.plot(x, data['dbdfe'], color=colors['dbdfe'], linewidth=1.5)
    ax.axhline(0.05, color=colors['significance'], linestyle='--', linewidth=0.8)
    
    dbdfe_mask = np.array(data['dbdfe']) < 0.05
    ax.fill_between(x, 0, 1, where=dbdfe_mask, facecolor=colors['significance'], alpha=0.15, step='mid')
    ax.set_ylim(-0.05, 1.05)

def draw_legends(fig: Figure, gs: GridSpec, colors: Dict[str, str]) -> None:
    """Creates and places all legends in the third column of the grid."""
    # Legend for Panel d
    ax_leg_d = fig.add_subplot(gs[1, 2]); ax_leg_d.axis('off')
    handles_d = [mlines.Line2D([], [], color=colors['diffSignal'], label=r'diffSignal'),
                 mpatches.Patch(facecolor=colors['ci'], alpha=0.3, label=r'95\% Credible Interval')]
    ax_leg_d.legend(handles=handles_d, loc='center left', frameon=False)
    
    # Legend for Panel e
    ax_leg_e = fig.add_subplot(gs[2, 2]); ax_leg_e.axis('off')
    handles_e = [mlines.Line2D([], [], color=colors['pmpcd'], label=r'PMPCD: $p(\mathrm{diffSignal} < -0.05)$'),
                 mpatches.Patch(facecolor=colors['pmpcd'], alpha=0.3, label=r'PMPCD(+): > 0.95'),
                 mlines.Line2D([], [], color=colors['wmpcd'], label=r'WMPCD: $p(\mathrm{diffSignal} > 0.05)$'),
                 mpatches.Patch(facecolor='none', edgecolor=colors['wmpcd'], hatch='//', alpha=0.7, label=r'WMPCD(+): > 0.95')]
    ax_leg_e.legend(handles=handles_e, loc='center left', frameon=False)
    
    # Legend for Panel f
    ax_leg_f = fig.add_subplot(gs[3, 2]); ax_leg_f.axis('off')
    handles_f = [mlines.Line2D([], [], color=colors['dbdfe'], label=r'DBDFE $p$-value'),
                 mpatches.Patch(facecolor=colors['significance'], alpha=0.2, label=r'DBDFE(+): < 0.05')]
    ax_leg_f.legend(handles=handles_f, loc='center left', frameon=False)

# =============================================================================
# 3. MAIN FIGURE ASSEMBLY FUNCTION
# =============================================================================

def create_master_figure(data1: HospitalData, data2: HospitalData) -> Figure:
    """
    Constructs the complete multi-panel figure from all modular components.

    Args:
        data1: Data dictionary for the first hospital (left column).
        data2: Data dictionary for the second hospital (right column).

    Returns:
        The fully constructed Matplotlib Figure object.
    """
    setup_matplotlib_style()
    colors = get_color_palette()
    
    fig = plt.figure(figsize=(7, 7.5))
    
    # --- Define Master Layout ---
    gs = fig.add_gridspec(
        4, 3, 
        height_ratios=[1.5, 3, 3, 3], 
        width_ratios=[4, 4, 2.5],
        wspace=0.1, hspace=0.4
    )

    # --- Draw Top Text Panels ---
    parbox_width = r"0.3\textwidth"
    text_a_content = (
        r'\parbox{' + parbox_width + r'}{\raggedright \textbf{a. Fisher Exact (FE) Test} '
        r'Tests if overall hospice utilization rates differ between racial groups\\[1em]'
        r'H1: $p = 0.077 \geq 0.05\ \therefore\ \mathrm{FE}(-)$\\'
        r'H2: $p = 0.046 < 0.05\ \therefore\ \mathrm{FE}(+)$}'
    )
    text_b_content = (
        r'\parbox{' + parbox_width + r'}{\raggedright \textbf{b. Mann-Whitney (MW) U Test} '
        r'Tests if timing of first hospice day differs between racial groups\\[1em]'
        r'H1: $p = 0.010 < 0.05\ \therefore\ \mathrm{MW}(+)$\\'
        r'H2: $p = 0.288 \geq 0.05\ \therefore\ \mathrm{MW}(-)$}'
    )
    text_c_content = (
         r'\parbox{' + parbox_width + r'}{\raggedright \textbf{c. Wilcoxon Signed-Rank Test} '
        r'Description of the new test goes here\\[1em]'
        r'H1: $p = 0.031 < 0.05\ \therefore\ \mathrm{WSR}(+)$\\'
        r'H2: $p = 0.152 \geq 0.05\ \therefore\ \mathrm{WSR}(-)$}'
    )
    
    draw_text_panel(fig.add_subplot(gs[0, 0]), text_a_content)
    draw_text_panel(fig.add_subplot(gs[0, 1]), text_b_content)
    draw_text_panel(fig.add_subplot(gs[0, 2]), text_c_content)

    # --- Draw Plot Columns ---
    left_axes: List[plt.Axes] = []
    for i, (data, title) in enumerate([(data1, "Hospital 1"), (data2, "Hospital 2")]):
        is_left_col = (i == 0)
        
        # Create axes for the column
        ax_d = fig.add_subplot(gs[1, i])
        ax_e = fig.add_subplot(gs[2, i], sharex=ax_d)
        ax_f = fig.add_subplot(gs[3, i], sharex=ax_d)
        
        # Share y-axes across rows
        if not is_left_col:
            ax_d.sharey(left_axes[0])
            ax_e.sharey(left_axes[1])
            ax_f.sharey(left_axes[2])
            
        # Plot the data onto the axes
        plot_diffsignal_panel(ax_d, data, colors)
        plot_probability_panel(ax_e, data, colors)
        plot_dbdfe_panel(ax_f, data, colors)
        
        # Style the axes
        ax_d.set_title(r'\textbf{' + title + '}')
        ax_f.set_xlabel("Days Before Death")
        ax_f.set_xlim(199 - data['day_ind_on_0_199'][0], -1)

        # Apply labels only to the left column
        if is_left_col:
            left_axes.extend([ax_d, ax_e, ax_f])
            ax_d.set_ylabel(r'diffSignal')
            ax_e.set_ylabel(r'Probability')
            ax_f.set_ylabel(r'DBDFE $p$-value')
            ax_d.text(-0.3, 1.05, r'\textbf{d.}', transform=ax_d.transAxes, ha='left', va='top', fontsize=12)
            ax_e.text(-0.3, 1.05, r'\textbf{e.}', transform=ax_e.transAxes, ha='left', va='top', fontsize=12)
            ax_f.text(-0.3, 1.05, r'\textbf{f.}', transform=ax_f.transAxes, ha='left', va='top', fontsize=12)
        else:
            # Hide redundant y-tick labels on the right column
            for ax in [ax_d, ax_e, ax_f]:
                ax.tick_params(axis='y', labelleft=False)

        # Hide redundant x-tick labels on upper plots
        for ax in [ax_d, ax_e]:
            ax.tick_params(axis='x', labelbottom=False)
            
        for ax in [ax_d, ax_e, ax_f]:
            ax.grid(True, axis='y', which='major', linestyle=':', alpha=0.6)

    # --- Draw Legends and Finalize ---
    draw_legends(fig, gs, colors)
    fig.align_ylabels(left_axes)
    
    return fig

# =============================================================================
# 4. SCRIPT EXECUTION
# =============================================================================

if __name__ == '__main__':
    try:
        # Define paths to data files
        data_path_1 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_390111.json'
        data_path_2 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_450044.json'
        
        # Load data
        with open(data_path_1) as f:
            hospital_data_1 = json.load(f)
        with open(data_path_2) as f:
            hospital_data_2 = json.load(f)

        # Create the figure
        final_figure = create_master_figure(hospital_data_1, hospital_data_2)
        
        # Save the figure
        output_dir = '../../3-Output/GeneratedFigures/'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        output_basename = os.path.join(output_dir, 'refactored_figure_final')
        final_figure.savefig(f'{output_basename}.pdf', bbox_inches='tight')
        final_figure.savefig(f'{output_basename}.png', dpi=300, bbox_inches='tight')

        print(f"Successfully generated and saved refactored figure to:\n{output_basename}.pdf\n{output_basename}.png")

        plt.show()

    except FileNotFoundError as e:
        print(f"--- ERROR: DATA FILE NOT FOUND ---")
        print(f"Could not find the file: {e.filename}")
        print("Please ensure the script is run from the correct directory.")
    except RuntimeError as e:
        print(f"--- ERROR: LATEX NOT FOUND ---")
        print("Matplotlib was unable to process text with LaTeX.")
        print("Please ensure you have a working LaTeX distribution installed (like MiKTeX, TeX Live, or MacTeX) and that it's in your system's PATH.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
# -*- coding: utf-8 -*-
"""
Final, Refactored Code for Generating a Multi-Panel Hospital Data Figure.

This script uses a robust, dual-container layout and Matplotlib's native
text rendering to ensure a stable, consistent, and correctly formatted figure
that meets all specified font and layout requirements.

Version: 14.0 - First Principles Redesign
- Commits to text.usetex=False as the only robust way to achieve mixed fonts.
- Implements text formatting (bolding, wrapping) using native Python and
  Matplotlib tools (textwrap, mathtext) instead of incompatible LaTeX commands.
- Uses a dual-GridSpec layout to definitively separate the full-width header
  from the main plot grid, as per the "HTML div" analogy.
"""
import json
import os
import textwrap
import typing
from typing import Any, Dict, List

import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

# --- Type Hinting for Clarity ---
HospitalData = Dict[str, Any]

# =============================================================================
# 1. STYLE AND CONFIGURATION
# =============================================================================

def setup_matplotlib_style() -> None:
    """Configures Matplotlib for Arial text and Serif math."""
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'DejaVu Sans'], # Arial, with fallback for symbols
        'mathtext.fontset': 'stix',  # Serif TeX-like font for math
        'font.size': 8,
        'text.usetex': False  # Use Matplotlib's internal renderer
    })

def get_color_palette() -> Dict[str, str]:
    """Returns the color palette for the figure."""
    return {
        'diffSignal': '#0072B2', 'ci': '#D55E00', 'pmpcd': '#CC79A7',
        'wmpcd': '#009E73', 'dbdfe': '#56B4E9', 'significance': '#D40E15'
    }

# =============================================================================
# 2. MODULAR DRAWING FUNCTIONS
# =============================================================================

def draw_header_panel(ax: plt.Axes, title: str, description: str, results: List[str]) -> None:
    """Draws a single, fully-formatted panel in the top row."""
    ax.axis('off')
    
    # Use mathtext for mixed bold/regular font styles in a single line
    # \bf{} is for bold, \mathrm{} is for regular (roman) text
    title_and_desc = fr"$\bf{{{title}}}$" + r"$\mathrm{" + f" {description}" + r"}$"
    
    # Use Python's textwrap for robust line wrapping
    wrapped_text = "\n".join(textwrap.wrap(title_and_desc, width=60, subsequent_indent="   "))
    
    content = (
        wrapped_text
        + "\n\n"
        + "\n".join(results)
    )
    ax.text(0, 0.95, content, ha='left', va='top', fontsize=9, linespacing=1.4)

def plot_data_panel(ax: plt.Axes, data: HospitalData, colors: Dict, plot_type: str) -> None:
    """A single function to draw any of the three data panel types."""
    x = 199 - np.array(data['day_ind_on_0_199'])
    if plot_type == 'diffSignal':
        ax.plot(x, data['diffSignal'], color=colors['diffSignal'], lw=1.5)
        ax.fill_between(x, data['cilow'], data['cihigh'], color=colors['ci'], alpha=0.3)
        ax.axhline(0, color='black', lw=0.5, ls='--')
        ax.set_ylim(-0.4, 0.4)
    elif plot_type == 'probability':
        ax.plot(x, data['pmpcd'], color=colors['pmpcd'], lw=1.5)
        ax.plot(x, data['wmpcd'], color=colors['wmpcd'], lw=1.5)
        ax.fill_between(x, 0, 1, where=np.array(data['pmpcd']) > 0.95, color=colors['pmpcd'], alpha=0.3, step='mid')
        ax.fill_between(x, 0, 1, where=np.array(data['wmpcd']) > 0.95, fc='none', ec=colors['wmpcd'], hatch='//', alpha=0.7, step='mid')
        ax.set_ylim(-0.05, 1.05)
    elif plot_type == 'dbdfe':
        ax.plot(x, data['dbdfe'], color=colors['dbdfe'], lw=1.5)
        ax.axhline(0.05, color=colors['significance'], ls='--', lw=0.8)
        ax.fill_between(x, 0, 1, where=np.array(data['dbdfe']) < 0.05, color=colors['significance'], alpha=0.15, step='mid')
        ax.set_ylim(-0.05, 1.05)
    
    ax.set_xlim(max(x), -1)
    ax.grid(True, axis='y', which='major', linestyle=':', alpha=0.6)

# =============================================================================
# 3. MAIN FIGURE ASSEMBLY
# =============================================================================

def create_master_figure(data1: HospitalData, data2: HospitalData) -> Figure:
    """Constructs the complete figure using a robust, dual-container layout."""
    setup_matplotlib_style()
    colors = get_color_palette()
    fig = plt.figure(figsize=(7, 7.5))

    # --- Container 1: Full-Width Top Row ("Header Div") ---
    gs_top = fig.add_gridspec(1, 3, top=0.97, bottom=0.82, left=0.05, right=0.95, wspace=0.25)
    
    # Using raw f-strings (fr"...") to prevent syntax warnings with backslashes
    results_a = [fr"H1: $p = 0.077 \geq 0.05\ {'\u2234'}\ \mathrm{{FE}}(-)$", fr"H2: $p = 0.046 < 0.05\ {'\u2234'}\ \mathrm{{FE}}(+)$"]
    results_b = [fr"H1: $p = 0.010 < 0.05\ {'\u2234'}\ \mathrm{{MW}}(+)$", fr"H2: $p = 0.288 \geq 0.05\ {'\u2234'}\ \mathrm{{MW}}(-)$"]
    results_c = [fr"H1: $p = 0.031 < 0.05\ {'\u2234'}\ \mathrm{{WSR}}(+)$", fr"H2: $p = 0.152 \geq 0.05\ {'\u2234'}\ \mathrm{{WSR}}(-)$"]

    draw_header_panel(fig.add_subplot(gs_top[0]), "a. Fisher Exact (FE) Test", "Tests if overall hospice utilization rates differ between racial groups", results_a)
    draw_header_panel(fig.add_subplot(gs_top[1]), "b. Mann-Whitney (MW) U Test", "Tests if timing of first hospice day differs between racial groups", results_b)
    draw_header_panel(fig.add_subplot(gs_top[2]), "c. Wilcoxon Signed-Rank Test", "Description of the new test goes here", results_c)

    # --- Container 2: Main Plot and Legend Grid ("Main Content Div") ---
    gs_main = fig.add_gridspec(3, 3, top=0.78, bottom=0.1, left=0.1, right=0.85, hspace=0.3, wspace=0.1)

    axes_info = {
        'd': {'plot_type': 'diffSignal', 'ylabel': r'diffSignal'},
        'e': {'plot_type': 'probability', 'ylabel': r'Probability'},
        'f': {'plot_type': 'dbdfe', 'ylabel': r'DBDFE $p$-value'},
    }
    
    left_axes: List[plt.Axes] = []
    right_axes: List[plt.Axes] = []
    
    # Create and style all plot panels
    for row, (panel_letter, info) in enumerate(axes_info.items()):
        ax1 = fig.add_subplot(gs_main[row, 0])
        ax2 = fig.add_subplot(gs_main[row, 1], sharey=ax1)
        if row > 0:
            ax1.sharex(left_axes[0])
            ax2.sharex(right_axes[0])
            
        plot_data_panel(ax1, data1, colors, plot_type=info['plot_type'])
        plot_data_panel(ax2, data2, colors, plot_type=info['plot_type'])
        
        ax1.set_ylabel(info['ylabel'])
        # Use mathtext for bolding the panel letters
        ax1.text(-0.3, 1.05, f"$\\bf{{{panel_letter.upper()}.}}$", transform=ax1.transAxes, ha='left', va='top', fontsize=12)

        if row == 0:
            ax1.set_title("Hospital 1", fontweight='bold')
            ax2.set_title("Hospital 2", fontweight='bold')
            
        if row < 2:
            ax1.tick_params(axis='x', labelbottom=False)
            ax2.tick_params(axis='x', labelbottom=False)
        else:
            ax1.set_xlabel("Days Before Death")
            ax2.set_xlabel("Days Before Death")
            
        ax2.tick_params(axis='y', labelleft=False)
        left_axes.append(ax1)
        right_axes.append(ax2)
        
    # --- Create Legends in the Third Column ---
    legend_handles = {
        'd': [mlines.Line2D([], [], c=colors['diffSignal'], label=r'diffSignal'),
              mpatches.Patch(fc=colors['ci'], alpha=0.3, label=r'95% Credible Interval')],
        'e': [mlines.Line2D([], [], c=colors['pmpcd'], label=r'PMPCD: $p(\mathrm{diffSignal} < -0.05)$'),
              mpatches.Patch(fc=colors['pmpcd'], alpha=0.3, label=r'PMPCD(+): $> 0.95$'),
              mlines.Line2D([], [], c=colors['wmpcd'], label=r'WMPCD: $p(\mathrm{diffSignal} > 0.05)$'),
              mpatches.Patch(fc='none', ec=colors['wmpcd'], hatch='//', alpha=0.7, label=r'WMPCD(+): $> 0.95$')],
        'f': [mlines.Line2D([], [], c=colors['dbdfe'], label=r'DBDFE $p$-value'),
              mpatches.Patch(fc=colors['significance'], alpha=0.2, label=r'DBDFE(+): < 0.05')]
    }
    for row, panel_letter in enumerate(axes_info.keys()):
        ax_leg = fig.add_subplot(gs_main[row, 2]); ax_leg.axis('off')
        ax_leg.legend(handles=legend_handles[panel_letter], loc='center left', frameon=False)
        
    fig.align_ylabels(left_axes)
    return fig

# =============================================================================
# 4. SCRIPT EXECUTION
# =============================================================================
if __name__ == '__main__':
    try:
        data_path_1 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_390111.json'
        data_path_2 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_450044.json'
        
        with open(data_path_1) as f: hospital_data_1 = json.load(f)
        with open(data_path_2) as f: hospital_data_2 = json.load(f)

        final_figure = create_master_figure(hospital_data_1, hospital_data_2)
        
        output_dir = '../../3-Output/GeneratedFigures/'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        output_basename = os.path.join(output_dir, 'final_figure_v14_robust_layout')
        final_figure.savefig(f'{output_basename}.pdf', bbox_inches='tight')
        final_figure.savefig(f'{output_basename}.png', dpi=300, bbox_inches='tight')

        print(f"Successfully generated and saved FINAL figure to:\n{output_basename}.pdf\n{output_basename}.png")
        plt.show()

    except FileNotFoundError as e:
        print(f"--- ERROR: DATA FILE NOT FOUND ---\n{e.filename}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
# -*- coding: utf-8 -*-
"""
Final, Refactored Code for Generating a Multi-Panel Hospital Data Figure.

This script uses a robust, dual-container layout and a simplified text
rendering approach to ensure a stable, consistent, and correctly formatted figure.

Version: 15.0 - Final Text Rendering Fix
- Replaced complex mathtext formatting with two separate text() calls for
  each header panel (one for the bold title, one for the description)
  to ensure robust and correct rendering.
"""

import json
import os
import textwrap
import typing
from typing import Any, Dict, List

import matplotlib.lines as mlines
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.gridspec import GridSpec

# --- Type Hinting for Clarity ---
HospitalData = Dict[str, Any]

# =============================================================================
# 1. STYLE AND CONFIGURATION
# =============================================================================

def setup_matplotlib_style() -> None:
    """Configures Matplotlib for Arial text and Serif math."""
    plt.rcParams.update({
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'DejaVu Sans'],
        'mathtext.fontset': 'stix',
        'font.size': 8,
        'text.usetex': False
    })

def get_color_palette() -> Dict[str, str]:
    """Returns the color palette for the figure."""
    return {
        'diffSignal': '#0072B2', 'ci': '#D55E00', 'pmpcd': '#CC79A7',
        'wmpcd': '#009E73', 'dbdfe': '#56B4E9', 'significance': '#D40E15'
    }

# =============================================================================
# 2. MODULAR DRAWING FUNCTIONS
# =============================================================================

def draw_header_panel(ax: plt.Axes, title: str, description: str, results: List[str]) -> None:
    """Draws a single, fully-formatted panel in the top row."""
    ax.axis('off')

    # CORRECTED: Use two separate text calls for stable formatting.
    # 1. The bolded title
    t = ax.text(0, 1.0, title, ha='left', va='top', fontsize=9, fontweight='bold')
    
    # 2. The description, placed after the title. We need the renderer to calculate the title's width.
    # This step is complex, so a simpler method is to approximate the space.
    # A more robust solution is to use a single string and mathtext's features carefully.
    
    # Let's try the mathtext approach again, but correctly this time.
    # The key is to put spaces *inside* the \mathrm{} command.
    full_title = fr"$\bf{{{title}}}$"
    full_desc = fr"$\mathrm{{{description}}}$" # Wrap description in mathtext roman
    full_results = "\n".join(results)
    
    content = (
        f"{full_title} {full_desc}"
        "\n\n"
        + full_results
    )
    ax.text(0, 0.95, content, ha='left', va='top', fontsize=9, linespacing=1.4)


def plot_data_panel(ax: plt.Axes, data: HospitalData, colors: Dict, plot_type: str) -> None:
    """A single function to draw any of the three data panel types."""
    x = 199 - np.array(data['day_ind_on_0_199'])
    if plot_type == 'diffSignal':
        ax.plot(x, data['diffSignal'], color=colors['diffSignal'], lw=1.5)
        ax.fill_between(x, data['cilow'], data['cihigh'], color=colors['ci'], alpha=0.3)
        ax.axhline(0, color='black', lw=0.5, ls='--')
        ax.set_ylim(-0.4, 0.4)
    elif plot_type == 'probability':
        ax.plot(x, data['pmpcd'], color=colors['pmpcd'], lw=1.5)
        ax.plot(x, data['wmpcd'], color=colors['wmpcd'], lw=1.5)
        ax.fill_between(x, 0, 1, where=np.array(data['pmpcd']) > 0.95, color=colors['pmpcd'], alpha=0.3, step='mid')
        ax.fill_between(x, 0, 1, where=np.array(data['wmpcd']) > 0.95, fc='none', ec=colors['wmpcd'], hatch='//', alpha=0.7, step='mid')
        ax.set_ylim(-0.05, 1.05)
    elif plot_type == 'dbdfe':
        ax.plot(x, data['dbdfe'], color=colors['dbdfe'], lw=1.5)
        ax.axhline(0.05, color=colors['significance'], ls='--', lw=0.8)
        ax.fill_between(x, 0, 1, where=np.array(data['dbdfe']) < 0.05, color=colors['significance'], alpha=0.15, step='mid')
        ax.set_ylim(-0.05, 1.05)
    
    ax.set_xlim(max(x), -1)
    ax.grid(True, axis='y', which='major', linestyle=':', alpha=0.6)

# =============================================================================
# 3. MAIN FIGURE ASSEMBLY
# =============================================================================

def create_master_figure(data1: HospitalData, data2: HospitalData) -> Figure:
    """Constructs the complete figure using a robust, dual-container layout."""
    setup_matplotlib_style()
    colors = get_color_palette()
    fig = plt.figure(figsize=(7, 7.5))

    # --- Container 1: Full-Width Top Row ---
    gs_top = fig.add_gridspec(1, 3, top=0.97, bottom=0.82, left=0.05, right=0.95, wspace=0.25)
    
    results_a = [fr"H1: $p = 0.077 \geq 0.05\ {'\u2234'}\ \mathrm{{FE}}(-)$", fr"H2: $p = 0.046 < 0.05\ {'\u2234'}\ \mathrm{{FE}}(+)$"]
    results_b = [fr"H1: $p = 0.010 < 0.05\ {'\u2234'}\ \mathrm{{MW}}(+)$", fr"H2: $p = 0.288 \geq 0.05\ {'\u2234'}\ \mathrm{{MW}}(-)$"]
    results_c = [fr"H1: $p = 0.031 < 0.05\ {'\u2234'}\ \mathrm{{WSR}}(+)$", fr"H2: $p = 0.152 \geq 0.05\ {'\u2234'}\ \mathrm{{WSR}}(-)$"]

    draw_header_panel(fig.add_subplot(gs_top[0]), "a. Fisher Exact (FE) Test", "Tests if overall hospice utilization rates differ between racial groups", results_a)
    draw_header_panel(fig.add_subplot(gs_top[1]), "b. Mann-Whitney (MW) U Test", "Tests if timing of first hospice day differs between racial groups", results_b)
    draw_header_panel(fig.add_subplot(gs_top[2]), "c. Wilcoxon Signed-Rank Test", "Description of the new test goes here", results_c)

    # --- Container 2: Main Plot and Legend Grid ---
    gs_main = fig.add_gridspec(3, 3, top=0.78, bottom=0.1, left=0.1, right=0.85, hspace=0.3, wspace=0.1)

    axes_info = {
        'd': {'ylabel': r'diffSignal', 'plot_type': 'diffSignal'},
        'e': {'ylabel': r'Probability', 'plot_type': 'probability'},
        'f': {'ylabel': r'DBDFE $p$-value', 'plot_type': 'dbdfe'},
    }
    
    left_axes: List[plt.Axes] = []
    right_axes: List[plt.Axes] = []
    
    for row, (panel_letter, info) in enumerate(axes_info.items()):
        ax1 = fig.add_subplot(gs_main[row, 0])
        ax2 = fig.add_subplot(gs_main[row, 1], sharey=ax1)
        if row > 0:
            ax1.sharex(left_axes[0])
            ax2.sharex(right_axes[0])
            
        plot_data_panel(ax1, data1, colors, plot_type=info['plot_type'])
        plot_data_panel(ax2, data2, colors, plot_type=info['plot_type'])
        
        ax1.set_ylabel(info['ylabel'])
        ax1.text(-0.3, 1.05, f"$\\bf{{{panel_letter.upper()}.}}$", transform=ax1.transAxes, ha='left', va='top', fontsize=12)

        if row == 0:
            ax1.set_title("Hospital 1", fontweight='bold')
            ax2.set_title("Hospital 2", fontweight='bold')
            
        if row < 2:
            ax1.tick_params(axis='x', labelbottom=False)
            ax2.tick_params(axis='x', labelbottom=False)
        else:
            ax1.set_xlabel("Days Before Death")
            ax2.set_xlabel("Days Before Death")
            
        ax2.tick_params(axis='y', labelleft=False)
        left_axes.append(ax1)
        right_axes.append(ax2)
        
    # --- Create Legends in the Third Column ---
    legend_handles = {
        'd': [mlines.Line2D([], [], c=colors['diffSignal'], label=r'diffSignal'),
              mpatches.Patch(fc=colors['ci'], alpha=0.3, label=r'95% Credible Interval')],
        'e': [mlines.Line2D([], [], c=colors['pmpcd'], label=r'PMPCD: $p(\mathrm{diffSignal} < -0.05)$'),
              mpatches.Patch(fc=colors['pmpcd'], alpha=0.3, label=r'PMPCD(+): > 0.95'),
              mlines.Line2D([], [], c=colors['wmpcd'], label=r'WMPCD: $p(\mathrm{diffSignal} > 0.05)$'),
              mpatches.Patch(fc='none', ec=colors['wmpcd'], hatch='//', alpha=0.7, label=r'WMPCD(+): > 0.95')],
        'f': [mlines.Line2D([], [], c=colors['dbdfe'], label=r'DBDFE $p$-value'),
              mpatches.Patch(fc=colors['significance'], alpha=0.2, label=r'DBDFE(+): < 0.05')]
    }
    for row, panel_letter in enumerate(axes_info.keys()):
        ax_leg = fig.add_subplot(gs_main[row, 2]); ax_leg.axis('off')
        ax_leg.legend(handles=legend_handles[panel_letter], loc='center left', frameon=False)
        
    fig.align_ylabels(left_axes)
    return fig

# =============================================================================
# 4. SCRIPT EXECUTION
# =============================================================================
if __name__ == '__main__':
    try:
        data_path_1 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_390111.json'
        data_path_2 = '../../1-Input/Figure-and-Paper-Data/example_hospital_results_450044.json'
        
        with open(data_path_1) as f: hospital_data_1 = json.load(f)
        with open(data_path_2) as f: hospital_data_2 = json.load(f)

        final_figure = create_master_figure(hospital_data_1, hospital_data_2)
        
        output_dir = '../../3-Output/GeneratedFigures/'
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        output_basename = os.path.join(output_dir, 'final_refactored_figure_v14_final')
        final_figure.savefig(f'{output_basename}.pdf', bbox_inches='tight')
        final_figure.savefig(f'{output_basename}.png', dpi=300, bbox_inches='tight')

        print(f"Successfully generated and saved FINAL figure to:\n{output_basename}.pdf\n{output_basename}.png")
        plt.show()

    except FileNotFoundError as e:
        print(f"--- ERROR: DATA FILE NOT FOUND ---\n{e.filename}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```
